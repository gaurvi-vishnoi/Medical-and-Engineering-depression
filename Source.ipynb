# Engineering Students

```python
train_set_eng,test_set_eng=train_test_split(Eng,test_size=0.2,random_state=0)
dep_columns=['Anxiety_Score','Depression_score','A_D']
ind_columns=['Age', 'Gender', 'Level_of_Study', 'Year_of_Study', 'Staying_in_hostel',
       'Type_of_Family', 'Family__Income_Level', 'Living_Place_Tier',
       'Sleeping_hours_(during_COVID)', 'Sleeping_pattern_(during_COVID)',
       'Sleep_Schedule_(during_COVID)', 'Tense_feeling', 'Frightened_feeling',
       'Worrying_thoughts', 'Can_feel_relaxed', 'Butterfly_Frightened_Feeling',
       'Restless_Feeling_COVID', 'Panic_Feeling_COVID',
       'Enjoy_Same_Things_As_COVID', 'Can_Laugh', 'Feel_cheerful',
       'Slowed_During_COVID', 'lost_interest_in_appearance_COVID',
       'Look_Forward_to_Things', 'Can_enjoy_during_COVID',
       'Online_Theory_Quality', 'Online_Practical_Quality',
       'Online_Learning_Quality', 'Online_Self_Learning_Quality',
       'Online_Teacher_Relationship_Quality', 'Online_Grades_Quality']
#Train
X_train_eng=train_set_eng[ind_columns]
y_train_eng_anx=train_set_eng['Anxiety_Score']
y_train_eng_dep=train_set_eng['Depression_score']
#Test
X_test_eng=test_set_eng[ind_columns]
y_test_eng_anx=test_set_eng['Anxiety_Score']
y_test_eng_dep=test_set_eng['Depression_score']
```

```python
from sklearn.metrics import mean_absolute_error
def model_tester(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    return mean_absolute_error(y_test,preds)
```

Anxiety

```python
#Regressor Testing
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDRegressor
from sklearn.svm import SVR
from sklearn.linear_model import BayesianRidge
from catboost import CatBoostRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.linear_model import LinearRegression
from xgboost.sklearn import XGBRegressor
from lightgbm import LGBMRegressor
model_1=RandomForestRegressor(n_estimators=100,random_state=0)
model_2=LinearRegression()
model_3=GradientBoostingRegressor(random_state=0)
model_4=ElasticNet(random_state=0)
model_5=SGDRegressor(random_state=0)
model_6=SVR()
model_7=BayesianRidge()
model_8=CatBoostRegressor(random_state=0)
model_9=KernelRidge()
model_10=XGBRegressor(random_state=0)
model_11=LGBMRegressor(random_state=0)
models=[model_1,model_2,model_3,model_4,model_5,model_6,model_7,model_8,model_9,model_10,model_11]
for i in models:
    print(f"{i} error = {model_tester(i,X_train_eng,y_train_eng_anx,X_test_eng,y_test_eng_anx)}")
```

```python
#Final anxiety prediction model for engineering students
model_eng_anx=GradientBoostingRegressor(random_state=0,learning_rate=0.06, max_depth=2, min_samples_split=7, n_estimators=500,
                                        subsample=0.4)
model_eng_anx.fit(X_train_eng,y_train_eng_anx)
preds_eng_anx=model_eng_anx.predict(X_test_eng)
print(mean_absolute_error(y_test_eng_anx,preds_eng_anx))
```

Depression

```python
#Model Testing
for i in models:
    print(f"Model = {i}, Error = {model_tester(i,X_train_eng,y_train_eng_dep,X_test_eng,y_test_eng_dep)}")
```

```python
#Final model for prediction of depression in engineering students
model_eng_dep=GradientBoostingRegressor(random_state=0,learning_rate=0.06, max_depth=2, min_samples_split=6, n_estimators=500, 
                                        subsample=0.6)
model_eng_dep.fit(X_train_eng,y_train_eng_dep)
preds_eng_dep=model_eng_dep.predict(X_test_eng)
print(mean_absolute_error(y_test_eng_dep,preds_eng_dep))
```

# Medical Students

```python
train_set_med,test_set_med=train_test_split(Meds,test_size=0.2,random_state=0)
#Train
X_train_med=train_set_med[ind_columns]
y_train_med_anx=train_set_med['Anxiety_Score']
y_train_med_dep=train_set_med['Depression_score']
#Test
X_test_med=test_set_med[ind_columns]
y_test_med_anx=test_set_med['Anxiety_Score']
y_test_med_dep=test_set_med['Depression_score']
```

###### Anxiety Model Testing

```python
#Regressor Testing
for i in models:
    print(f"{i} error = {model_tester(i,X_train_med,y_train_med_anx,X_test_med,y_test_med_anx)}",end='\n''\n')
```

```python
#Final anxiety prediction model for medical students
model_med_anx=ElasticNet(random_state=0, alpha=0.1, l1_ratio=0.5, max_iter=500)
model_med_anx.fit(X_train_med,y_train_med_anx)
preds_med_anx=model_med_anx.predict(X_test_med)
print(mean_absolute_error(y_test_med_anx,preds_med_anx))
```

```python
#Same model used for depression analysis in Med students aswell due to small dataset
model_med_dep=ElasticNet(random_state=0, alpha=1, l1_ratio=0.5, max_iter=500)
model_med_dep.fit(X_train_med,y_train_med_dep)
preds_dep_med=model_med_dep.predict(X_test_med)
print(mean_absolute_error(y_test_med_dep,preds_dep_med))
```

## Hyperparameter Tuning
